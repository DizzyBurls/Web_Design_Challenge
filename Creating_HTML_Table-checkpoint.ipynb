## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the total number of schools in the district.

unique_schools=school_data_merged_df["school_name"].unique()
number_schools=len(unique_schools)
number_schools
```

```python
# Calculate the total number of students in the district.

unique_students=school_data_merged_df["Student ID"].unique()
number_students=len(unique_students)
number_students
```

```python
# Calculate the average Maths score in the district.

average_maths=school_data_merged_df["math_score"].mean()
average_maths
```

```python
# Calculate the average Reading score in the district.

average_reading=school_data_merged_df["reading_score"].mean()
average_reading
```

```python
# Compile a list of all the different schools budget figures for the district.

unique_budget=school_data_merged_df["budget"].unique()
unique_budget
```

```python
# Check to ensure that there are no duplicate budget figures on the list. 

if len(unique_budget)==number_schools:
    print("All schools have a unique budget figure. All budget calcutions will be correct from this point.")
else:
    print("There is one or more repeated budget figures in the merged data frame. All repeated budget figures must be added to 'Total District Budget' figure below.")
```

```python
# Determine the Total District Budget:

total_budget=unique_budget.sum()
total_budget
```

```python
#Determine which students scored 70 or higher in Maths.

passing_maths_df=school_data_merged_df.loc[school_data_merged_df["math_score"] >= 70, ["budget","type","Student ID", "math_score", "school_name", "grade"]]

# Determine how many students scored 70 or higher in Maths.

unique_passing_maths=passing_maths_df["Student ID"].unique()
number_passing_maths=len(unique_passing_maths)

# Determine the percentage of students who scored 70 or higher in Maths.

percentage_passing_maths=100*number_passing_maths/number_students
percentage_passing_maths
```

```python
#Determine which students scored 70 or higher in Reading.

passing_reading_df=school_data_merged_df.loc[school_data_merged_df["reading_score"] >= 70, ["Student ID", "math_score", "school_name", "grade"]]

# Determine how many students scored 70 or higher in Reading.

unique_passing_reading=passing_reading_df["Student ID"].unique()
number_passing_reading=len(unique_passing_reading)

# Determine the percentage of students who scored 70 or higher in Reading.

percentage_passing_reading=100*number_passing_reading/number_students
percentage_passing_reading
```

```python
#Determine which students scored 70 or higher in Reading and Maths.

passing_both_df=passing_reading_df.loc[passing_reading_df["math_score"] >= 70, ["Student ID", "school_name"]]

# Determine how many students scored 70 or higher in Reading and Maths.

unique_passing_both=passing_both_df["Student ID"].unique()
number_passing_both=len(unique_passing_both)

# Determine the percentage of students who scored 70 or higher in Reading and Maths.

percentage_passing_both=100*number_passing_both/number_students
percentage_passing_both
```

```python
#Create a new Data Frame with the appropriate column headings:

district_summary_df = pd.DataFrame([{"Number of Schools":number_schools,"Number of Students":number_students,"Budget":total_budget,"Average Maths Score":average_maths,"Average Reading Score":average_reading,"% Passing Maths":percentage_passing_maths,"% Passing Reading":percentage_passing_reading,"% Pass Maths and Reading":percentage_passing_both}])

#Format currency columns:

district_summary_df["Budget"]=district_summary_df["Budget"].map("${:,.2f}".format)
district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Create a Data Frame called "Schools_Summary_df".

Schools_Summary_df=pd.DataFrame()

# Add the unique_school list as elements in the first column, "School Name".

Schools_Summary_df['school_name']=unique_schools

# Sort the School Names alphabetically.

Schools_Summary_df.sort_values(by='school_name', inplace=True)

Schools_Summary_df
```

```python
# Determine each school's reading average.

reading_average=school_data_merged_df.groupby(['school_name'])['reading_score'].mean()

# Convert it to a data frame for future merging.

reading_average_df=pd.DataFrame({"reading_average": reading_average})
reading_average_df
```

```python
# Determine each school's maths average.

maths_average=school_data_merged_df.groupby(['school_name'])['math_score'].mean()

# Convert it to a data frame for future merging.

maths_average_df=pd.DataFrame({"maths_average": maths_average})
maths_average_df
```

```python
# Determine each schools Enrolment count.

school_enrolments=school_data_merged_df.groupby(['school_name'])["Student ID"].count()

# Convert it to a data frame for future merging.

school_enrolments_df=pd.DataFrame({"school_enrolments": school_enrolments})
school_enrolments_df
```

```python
# Group the list of students passing maths by School Name.

group_passing_maths=passing_maths_df.groupby(['school_name'])

# Determine the size of each group.

school_students_passing_maths=group_passing_maths.size()

# Convert it to a data frame for column calculation.

school_students_passing_maths_df=pd.DataFrame({"school_students_passing_maths": school_students_passing_maths})
school_students_passing_maths_df

# Convert it to a percentage.

percentage_school_maths=(school_students_passing_maths_df["school_students_passing_maths"])*100/(school_enrolments_df["school_enrolments"])
#percentage_school_maths

# Convert it to a data frame for future merging.

percentage_school_maths_df=pd.DataFrame({"percentage_school_maths": percentage_school_maths})
percentage_school_maths_df
```

```python
# Group the list of students passing reading by School Name.

group_passing_reading=passing_reading_df.groupby(['school_name'])

# Determine the size of each group.

school_students_passing_reading=group_passing_reading.size()

# Convert it to a data frame for column calculation.

school_students_passing_reading_df=pd.DataFrame({"school_students_passing_reading": school_students_passing_reading})
school_students_passing_reading_df

# Convert it to a percentage.

percentage_school_reading=(school_students_passing_reading_df["school_students_passing_reading"])*100/(school_enrolments_df["school_enrolments"])
#percentage_school_reading

# Convert it to a data frame for future merging.

percentage_school_reading_df=pd.DataFrame({"percentage_school_reading": percentage_school_reading})
percentage_school_reading_df
```

```python
# Group the list of students passing both by School Name.

group_passing_both=passing_both_df.groupby(['school_name'])

school_students_passing_both=group_passing_both.size()

# Convert it to a data frame for column calculation.

school_students_passing_both_df=pd.DataFrame({"school_students_passing_both": school_students_passing_both})
school_students_passing_both_df

# Convert it to a percentage.

percentage_school_both=(school_students_passing_both_df["school_students_passing_both"])*100/(school_enrolments_df["school_enrolments"])
#percentage_school_reading

# Convert it to a data frame for future merging.

percentage_school_both_df=pd.DataFrame({"percentage_school_both": percentage_school_both})
percentage_school_both_df
```

```python
# Determine each School Type.

school_type_df=group_passing_maths.first()

# Delete unneeded columns from Data Frame.

del school_type_df['Student ID']
del school_type_df['math_score']
del school_type_df['budget']
del school_type_df['grade']

school_type_df
```

```python
# Determine each School Budget.

school_budget_df=group_passing_maths.first()

# Delete unneeded columns from Data Frame.

del school_budget_df['Student ID']
del school_budget_df['math_score']
del school_budget_df['type']
del school_budget_df['grade']

school_budget_df
```

```python
# Determine the school Budget per student.

school_budget_per_student=(school_budget_df["budget"])/(school_enrolments_df["school_enrolments"])

# Convert it to a Data Frame for future merge.

school_budget_per_student_df=pd.DataFrame({"school_budget_per_student": school_budget_per_student})
school_budget_per_student_df
```

```python
# Merge all relevent Data Frames.

Schools_Summary_df1=pd.merge(Schools_Summary_df, reading_average_df, how="left", on=["school_name"])
Schools_Summary_df2=pd.merge(Schools_Summary_df1, maths_average_df, how="left", on=["school_name"])
Schools_Summary_df3=pd.merge(Schools_Summary_df2, school_enrolments_df, how="left", on=["school_name"])
Schools_Summary_df4=pd.merge(Schools_Summary_df3, percentage_school_maths_df, how="left", on=["school_name"])
Schools_Summary_df5=pd.merge(Schools_Summary_df4, percentage_school_reading_df, how="left", on=["school_name"])
Schools_Summary_df6=pd.merge(Schools_Summary_df5, school_type_df, how="left", on=["school_name"])
Schools_Summary_df7=pd.merge(Schools_Summary_df6, school_budget_df, how="left", on=["school_name"])
Schools_Summary_df8=pd.merge(Schools_Summary_df7, school_budget_per_student_df, how="left", on=["school_name"])
Schools_Summary_df9=pd.merge(Schools_Summary_df8, percentage_school_both_df, how="left", on=["school_name"])

# Rename Columns.

Schools_Summary_df10=Schools_Summary_df9.rename(columns={"school_name": "School", "reading_average": "Average Reading Score", "maths_average": "Average Math Score", "budget": "Total School Budget", "type": "School Type", "school_enrolments": "Total Students", "percentage_school_reading":"% Passing Reading", "percentage_school_maths": "% Passing Maths", "school_budget_per_student": "Per Student Budget", "percentage_school_both": "% Overall Passing"})

# Organise Columns.

Master_Schools_Summary_Index_df=Schools_Summary_df10[["School", "School Type", "Total Students", "Total School Budget", "Per Student Budget", "Average Math Score", "Average Reading Score", "% Passing Maths", "% Passing Reading", "% Overall Passing"]]

# Set index to School.

Master_Schools_Summary_df=Master_Schools_Summary_Index_df.set_index("School")

# Format currency columns.

Master_Schools_Summary_Formatted_df=Master_Schools_Summary_df
Master_Schools_Summary_Formatted_df["Per Student Budget"]=Master_Schools_Summary_Formatted_df["Per Student Budget"].map("${:,.2f}".format)
Master_Schools_Summary_Formatted_df["Total School Budget"]=Master_Schools_Summary_Formatted_df["Total School Budget"].map("${:,.2f}".format)
Master_Schools_Summary_Formatted_df
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Create a series for the Year 9 Data for Maths Scores:
    
Year9_Level_df=school_data_merged_df.loc[school_data_merged_df["grade"] == "9th", ["school_name", "math_score", "grade"]]
Year9_Level_df=Year9_Level_df.rename(columns={"math_score": "9 Maths Score"})

# Group the Year 9 Maths Scores data by School, and then take the average of that grouped data for each school.

average_9maths_school_grade_df=Year9_Level_df.groupby(['school_name']).mean()

# Create a series for the Year 10 Data for Maths Scores:

Year10_Level_df=school_data_merged_df.loc[school_data_merged_df["grade"] == "10th", ["school_name", "math_score", "grade"]]
Year10_Level_df=Year10_Level_df.rename(columns={"math_score": "10 Maths Score"})

# Group the Year 10 Maths Scores data by School, and then take the average of that grouped data for each school.

average_10maths_school_grade_df=Year10_Level_df.groupby(['school_name']).mean()

# Create a series for the Year 11 Data for Maths Scores:

Year11_Level_df=school_data_merged_df.loc[school_data_merged_df["grade"] == "11th", ["school_name", "math_score", "grade"]]
Year11_Level_df=Year11_Level_df.rename(columns={"math_score": "11 Maths Score"})

# Group the Year 11 Maths Scores data by School, and then take the average of that grouped data for each school.

average_11maths_school_grade_df=Year11_Level_df.groupby(['school_name']).mean()

# Create a series for the Year 12 Data for Maths Scores:

Year12_Level_df=school_data_merged_df.loc[school_data_merged_df["grade"] == "12th", ["school_name", "math_score", "grade"]]
Year12_Level_df=Year12_Level_df.rename(columns={"math_score": "12 Maths Score"})

# Group the Year 12 Maths Scores data by School, and then take the average of that grouped data for each school.

average_12maths_school_grade_df=Year12_Level_df.groupby(['school_name']).mean()

# Merge all of the relevent data frames together.

Year_Level_Breakdown_df1=pd.merge(average_9maths_school_grade_df, average_10maths_school_grade_df, how="left", on=["school_name"])
Year_Level_Breakdown_df2=pd.merge(Year_Level_Breakdown_df1, average_11maths_school_grade_df, how="left", on=["school_name"])
Year_Level_Breakdown_df3=pd.merge(Year_Level_Breakdown_df2, average_12maths_school_grade_df, how="left", on=["school_name"])

# Rename the column headers.

Year_Level_Breakdown_df3=Year_Level_Breakdown_df3.rename(columns={"9 Maths Score": "9th", "10 Maths Score": "10th", "11 Maths Score": "11th", "12 Maths Score": "12th"})
Year_Level_Breakdown_df3
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Ensure that data frame is unformatted version.

Master_Schools_Summary_df=Master_Schools_Summary_Index_df.set_index("School")

# Create bins for the data based on a school's Budget Per Student and label those bins.

bins=[0, 585, 630, 645, 680]
labels=['<$585', '$585-630', '$630-645','$645-680']

# Bin the Maths Score data and find the average of each bin. Create a data frame of each bin's average.

Master_Schools_Summary_df['Spending Ranges (Per Student)']=pd.cut(Master_Schools_Summary_df['Per Student Budget'], bins=bins, labels=labels)
Bins_Spend_AvMaths=Master_Schools_Summary_df.groupby(['Spending Ranges (Per Student)'])['Average Math Score'].mean()
Bins_Spend_AvMaths_df=pd.DataFrame({"Bins_Spend_AvMaths":Bins_Spend_AvMaths})

# Bin the Reading Score data and find the average of each bin. Create a data frame of each bin's average.

Bins_Spend_AvReading=Master_Schools_Summary_df.groupby(['Spending Ranges (Per Student)'])['Average Reading Score'].mean()
Bins_Spend_AvReading_df=pd.DataFrame({"Bins_Spend_AvReading":Bins_Spend_AvReading})

# Bin the % Passing Maths data and find the average of each bin. Create a data frame of each bin's average.

Bins_Spend_PassingMaths=Master_Schools_Summary_df.groupby(['Spending Ranges (Per Student)'])['% Passing Maths'].mean()
Bins_Spend_PassingMaths_df=pd.DataFrame({"Bins_Spend_PassingMaths": Bins_Spend_PassingMaths})

# Bin the % Passing Reading data and find the average of each bin. Create a data frame of each bin's average.

Bins_Spend_PassingReading=Master_Schools_Summary_df.groupby(['Spending Ranges (Per Student)'])['% Passing Reading'].mean()
Bins_Spend_PassingReading_df=pd.DataFrame({"Bins_Spend_PassingReading": Bins_Spend_PassingReading})

# Bin the % Passing Overall data and find the average of each bin. Create a data frame of each bin's average.

Bins_Spend_PassingOverall=Master_Schools_Summary_df.groupby(['Spending Ranges (Per Student)'])['% Overall Passing'].mean()
Bins_Spend_PassingOverall_df=pd.DataFrame({"Bins_Spend_PassingOverall": Bins_Spend_PassingOverall})

# Merge the relevent data frames.

Bins_Spend_Master_df1=pd.merge(Bins_Spend_AvMaths_df, Bins_Spend_AvReading_df, how="left", on=['Spending Ranges (Per Student)'])
Bins_Spend_Master_df2=pd.merge(Bins_Spend_Master_df1, Bins_Spend_PassingMaths_df, how="left", on=['Spending Ranges (Per Student)'])
Bins_Spend_Master_df3=pd.merge(Bins_Spend_Master_df2, Bins_Spend_PassingReading_df, how="left", on=['Spending Ranges (Per Student)'])
Bins_Spend_Master_df4=pd.merge(Bins_Spend_Master_df3, Bins_Spend_PassingOverall_df, how="left", on=['Spending Ranges (Per Student)'])

# Rename the column headers.

Bins_Spend_Master_df5=Bins_Spend_Master_df4.rename(columns={"Bins_Spend_AvMaths": "Average Math Score", "Bins_Spend_AvReading": "Average Reading Score", "Bins_Spend_PassingMaths": "% Passing Math", "Bins_Spend_PassingReading": "% Passing Reading", "Bins_Spend_PassingOverall": "% Passing Overall", })

# Format the columns to match solutions.

Bins_Spend_Master_df5["Average Math Score"]=Bins_Spend_Master_df5["Average Math Score"].map("{:.2f}".format)
Bins_Spend_Master_df5["Average Reading Score"]=Bins_Spend_Master_df5["Average Reading Score"].map("{:.2f}".format)
Bins_Spend_Master_df5["% Passing Math"]=Bins_Spend_Master_df5["% Passing Math"].map("{:.2f}".format)
Bins_Spend_Master_df5["% Passing Reading"]=Bins_Spend_Master_df5["% Passing Reading"].map("{:.2f}".format)
Bins_Spend_Master_df5["% Passing Overall"]=Bins_Spend_Master_df5["% Passing Overall"].map("{:.2f}".format)

Bins_Spend_Master_df5
```

```python
# Create bins for the data based on a school's Total Students and label those bins.

bins=[0, 1000, 2000, 5000]
labels=['Small <1000', 'Medium 1000-2000', 'Large 2000-5000']

# Bin the Maths Score data and find the average of each bin. Create a data frame of each bin's average.

Master_Schools_Summary_df['School Size']=pd.cut(Master_Schools_Summary_df['Total Students'], bins=bins, labels=labels)
Bins_Size_AvMaths=Master_Schools_Summary_df.groupby(['School Size'])['Average Math Score'].mean()
Bins_Size_AvMaths_df=pd.DataFrame({"Bins_Size_AvMaths":Bins_Size_AvMaths})

# Bin the Reading Score data and find the average of each bin. Create a data frame of each bin's average.

Bins_Size_AvReading=Master_Schools_Summary_df.groupby(['School Size'])['Average Reading Score'].mean()
Bins_Size_AvReading_df=pd.DataFrame({"Bins_Size_AvReading":Bins_Size_AvReading})

# Bin the % Passing Maths data and find the average of each bin. Create a data frame of each bin's average.

Bins_Size_PassingMaths=Master_Schools_Summary_df.groupby(['School Size'])['% Passing Maths'].mean()
Bins_Size_PassingMaths_df=pd.DataFrame({"Bins_Size_PassingMaths": Bins_Size_PassingMaths})

# Bin the % Passing Reading data and find the average of each bin. Create a data frame of each bin's average.

Bins_Size_PassingReading=Master_Schools_Summary_df.groupby(['School Size'])['% Passing Reading'].mean()
Bins_Size_PassingReading_df=pd.DataFrame({"Bins_Size_PassingReading": Bins_Size_PassingReading})

# Bin the % Passing Overall data and find the average of each bin. Create a data frame of each bin's average.

Bins_Size_PassingOverall=Master_Schools_Summary_df.groupby(['School Size'])['% Overall Passing'].mean()
Bins_Size_PassingOverall_df=pd.DataFrame({"Bins_Size_PassingOverall": Bins_Size_PassingOverall})


# Merge the relevent data frames.

Bins_Size_Master_df1=pd.merge(Bins_Size_AvMaths_df, Bins_Size_AvReading_df, how="left", on=['School Size'])
Bins_Size_Master_df2=pd.merge(Bins_Size_Master_df1, Bins_Size_PassingMaths_df, how="left", on=['School Size'])
Bins_Size_Master_df3=pd.merge(Bins_Size_Master_df2, Bins_Size_PassingReading_df, how="left", on=['School Size'])
Bins_Size_Master_df4=pd.merge(Bins_Size_Master_df3, Bins_Size_PassingOverall_df, how="left", on=['School Size'])

# Rename the column headers.

Bins_Size_Master_df5=Bins_Size_Master_df4.rename(columns={"Bins_Size_AvMaths": "Average Math Score", "Bins_Size_AvReading": "Average Reading Score", "Bins_Size_PassingMaths": "% Passing Math", "Bins_Size_PassingReading": "% Passing Reading", "Bins_Size_PassingOverall": "% Passing Overall", })
Bins_Size_Master_df5
```
